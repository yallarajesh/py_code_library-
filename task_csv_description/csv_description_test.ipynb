```python
import sys
sys.path.insert(0, '.')
from csv_description import (
    read_csv_file,
    get_basic_info,
    get_dtypes,
    get_missing_values,
    get_unique_counts,
    get_numeric_summary,
    get_categorical_summary,
    get_memory_usage,
    get_sample_rows,
    describe_csv,
)

sample_path = '/home/claude/test_data/sample.csv'
```

```python
df = read_csv_file(sample_path)
assert df.shape == (6, 5)
print(df)
```

```python
basic_info = get_basic_info(df)
assert basic_info['row_count'] == 6
assert basic_info['column_count'] == 5
assert basic_info['column_names'] == ['id', 'name', 'age', 'city', 'score']
basic_info
```

```python
dtypes = get_dtypes(df)
assert set(dtypes.keys()) == set(df.columns)
dtypes
```

```python
missing = get_missing_values(df)
assert missing['age']['missing_count'] == 1
assert missing['score']['missing_count'] == 1
assert missing['id']['missing_count'] == 0
missing
```

```python
unique_counts = get_unique_counts(df)
assert unique_counts['city'] == 3
assert unique_counts['id'] == 6
unique_counts
```

```python
numeric_summary = get_numeric_summary(df)
assert 'age' in numeric_summary
assert 'score' in numeric_summary
assert 'name' not in numeric_summary
numeric_summary
```

```python
categorical_summary = get_categorical_summary(df)
assert 'city' in categorical_summary
assert 'name' in categorical_summary
assert categorical_summary['city']['Hyderabad'] == 2
categorical_summary
```

```python
memory_usage = get_memory_usage(df)
assert memory_usage['total_bytes'] > 0
assert set(memory_usage['per_column_bytes'].keys()) == set(df.columns)
memory_usage
```

```python
sample_rows = get_sample_rows(df, sample_size=3)
assert len(sample_rows) == 3
assert sample_rows[0]['name'] == 'Alice'
sample_rows
```

```python
full_description = describe_csv(sample_path, sample_size=2)
expected_keys = {
    'basic_info', 'dtypes', 'missing_values', 'unique_counts',
    'numeric_summary', 'categorical_summary', 'memory_usage', 'sample_rows'
}
assert set(full_description.keys()) == expected_keys
assert len(full_description['sample_rows']) == 2
full_description
```
